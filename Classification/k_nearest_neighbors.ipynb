````
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
````

Excellent observation, Tushar! You're touching on a **very important best practice in machine learning**  how to correctly scale your data for models like SVM, KNN, Logistic Regression, etc.

Let's break down **why we do `fit_transform` on `X_train`**, but only `transform` on `X_test`.

---

##  First, Quick Recap

###  `fit_transform()`

* Learns the **mean** and **standard deviation** from the data
* Then applies standardization:

  $$
  X_{\text{scaled}} = \frac{X - \mu}{\sigma}
  $$

###  `transform()`

* Uses **existing** mean and std (already learned)
* Applies the same transformation

---

##  The Correct Approach:

```python
sc = StandardScaler()
X_train = sc.fit_transform(X_train)  #  Learn mean and std from training set
X_test = sc.transform(X_test)        #  Use the same mean and std
```

---

##  Why Not `fit_transform()` on `X_test`?

Because it would **cheat**.

If you do:

```python
X_test = sc.fit_transform(X_test)  #  Wrong!
```

...you're calculating a **new mean and std** from the test set  which the model **shouldn't know** during training.

That introduces **data leakage**, which:

* Gives you **unrealistically good results**
* Violates the principle of keeping test data **unseen**

---

##  Think of It Like This:

* `X_train`  used to **train and fit the scaler**
* `X_test`  must be **transformed using the same rules** (mean & std) that the model learned during training

You want to evaluate:

> How does my model perform on **new, unseen data** that follows the same scale as training?

---

##  Summary

| Step                     | Purpose                            | Use   |
| ------------------------ | ---------------------------------- | ----- |
| `fit_transform(X_train)` | Learn scaling + apply it           |  Yes |
| `transform(X_test)`      | Use learned scaling from train     |  Yes |
| `fit_transform(X_test)`  | Learns new scaling (data leakage!) |  No  |

## Training the K-NN model on the Training set

````
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)
classifier.fit(X_train, y_train)
````

Great! You're now training a **K-Nearest Neighbors (KNN) Classifier** using scikit-learn. Let me explain this line **step-by-step** so you understand exactly what each part means and does:

---

##  Full Code

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors=5, metric='minkowski', p=2)
classifier.fit(X_train, y_train)
```

---

##  Step-by-Step Explanation

---

### 1 `from sklearn.neighbors import KNeighborsClassifier`

* This imports the **KNN Classifier** from scikit-learn.
* It's used for **classification tasks** (e.g., yes/no, spam/ham, cat/dog).

---

### 2 `KNeighborsClassifier(...)`

This initializes the KNN model with specific parameters:

####  `n_neighbors=5`

* This sets **k = 5**  i.e., use the **5 nearest neighbors** to make predictions.
* The class predicted will be the **majority vote** among these 5.

####  `metric='minkowski'` and `p=2`

* KNN uses **distance** to measure nearness.

* `'minkowski'` is a **general distance formula**:

  $$
  D = \left( \sum |x_i - y_i|^p \right)^{1/p}
  $$

* When `p=2`, this becomes:

  $$
  D = \sqrt{(x_1 - y_1)^2 + (x_2 - y_2)^2 + ...}
  $$

 Thats just **Euclidean distance** (straight-line distance).

> TL;DR: you're telling KNN to use **Euclidean distance** to find the 5 closest training samples.

---

### 3 `classifier.fit(X_train, y_train)`

* This trains the model  although technically:

  * KNN doesnt **learn weights or coefficients**.
  * It just **stores the training data** so it can use it during prediction.

Thats why KNN is called a **lazy learner**.

---

##  Visual Summary

| Parameter                 | Meaning                                             |
| ------------------------- | --------------------------------------------------- |
| `n_neighbors=5`           | Use 5 closest points to vote                        |
| `metric='minkowski', p=2` | Use Euclidean distance                              |
| `.fit(...)`               | Stores training data for future distance comparison |

## Predicting the Test set results

````
y_pred = classifier.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
````

## Making the Confusion Matrix

````
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
````

## Visualising the Training set results

````
from matplotlib.colors import ListedColormap
X_set, y_set = sc.inverse_transform(X_train), y_train
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 10, stop = X_set[:, 0].max() + 10, step = 1),
                     np.arange(start = X_set[:, 1].min() - 1000, stop = X_set[:, 1].max() + 1000, step = 1))
plt.contourf(X1, X2, classifier.predict(sc.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('K-NN (Training set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
````
